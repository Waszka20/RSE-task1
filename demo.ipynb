```python
from fasttrajectory import simulate_projectile

x, y = simulate_projectile(
    velocity=90,
    angle_deg=60,
    drag=0.01
)

print(f"Range: {x[-1]:.2f} m")
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y)
plt.xlabel("Distance (m)")
plt.ylabel("Height (m)")
plt.title("Projectile Motion")
plt.grid(True)

plt.show()
```
